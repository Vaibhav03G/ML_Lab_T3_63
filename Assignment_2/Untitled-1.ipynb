```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
# Get dataset
df_sal = pd.read_csv('C:\\IMP\\github\\ML_Lab_T3_63\\Assignment_2\\Student_Marks.csv')
df_sal.head()
```

```python
# Describe data
df_sal.describe()
```

```python
# Data distribution
plt.title('Marks Distribution Plot')
sns.distplot(df_sal['Marks'])
plt.show()
```

```python
# Relationship between Salary and Experience
plt.scatter(df_sal['time_study'], df_sal['Marks'], color = 'lightcoral')
plt.title('Marks vs time_study')
plt.xlabel('Years of time_study')
plt.ylabel('Marks')
plt.box(False)
plt.show()
```

```python
# Splitting variables
X = df_sal.iloc[:, :1]  # independent
y = df_sal.iloc[:, 1:]  # dependent
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Regressor model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Prediction result
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)   # predicted value of y_train
```

```python
# Regressor coefficients and intercept
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```

```python
X = df_sal[ 'time_study' ]
y = df_sal[ 'Marks' ]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, train_size = 0.7, test_size = 0.3, random_state = 100 )
```

```python
print( X_train.shape )
print( X_test.shape )
print( y_train.shape )
print( y_test.shape )
```

```python
import statsmodels.api as sm
```

```python
# Add a constant to get an intercept
X_train_sm = sm.add_constant(X_train)
# Fit the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()
```

```python
# Print the parameters,i.e. intercept and slope of the regression line obtained
lr.params
```

```python
#Performing a summary operation lists out all different parameters of the regression line fitted
print(lr.summary())
```

```python
# Add a constant to X_test
X_test_sm = sm.add_constant(X_test)
# Predict the y values corresponding to X_test_sm
y_pred = lr.predict(X_test_sm)
```

```python
y_pred.head()
```

```python
#Imporitng libraries
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
#RMSE value
print("RMSE: ",np.sqrt( mean_squared_error( y_test, y_pred ) ))
#R-squared value
print( "R-squared: ",r2_score( y_test, y_pred ) )
```
